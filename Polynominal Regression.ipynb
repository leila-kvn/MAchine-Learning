```python
%matplotlib inline
from pylab import *
import numpy as np
np.random.seed(4)
x=np.random.normal(3,1,1000)
y=np.random.normal(50,10,1000)/x
scatter(x,y)
```

```python
import matplotlib.pyplot as plt

a=np.array(x)
b=np.array(y)
p9=np.poly1d(np.polyfit(a,b,9))
xp=np.linspace(0,6,300)
plt.scatter(x,y)
plt.plot(xp,p9(xp),c='b')
plt.show()
```
