```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.random.uniform(0,10,100)
X= np.sort(X)
noise=np.random.normal(0,0.3,100)
y=np.sin(X)+ noise- 0.2* X
X= X[:,np.newaxis]
plt.scatter(X,y)
X_train,X_test,y_train,y_test= train_test_split(X,y,train_size=0.8)
X_train.shape, y_train.shape
```

```python
lin_reg= LinearRegression()
lin_reg.fit(X_train,y_train)
lin_reg.coef_
y_prd=lin_reg.predict(X_test)
accuracy=lin_reg.score(X_test,y_test)
accuracy2=mean_squared_error(y_test, y_prd)
```

```python
def visualize_regression(x,y,model):
    plt.scatter(x,y,color='gray')
    r_line=np.linspace(0,10,100)
    r_line=r_line[:,np.newaxis]
    plt.plot(r_line,model.predict(r_line))
visualize_regression(X_train,y_train,lin_reg)    
```
